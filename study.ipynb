```python
import copy

a = ["주", "석", "티노"]

def Myappend(n):
  a.append(n)
  return a

def Mypop():
  x = a[len(a)-1]
  del(a[len(a)-1])
  return a

def Myreverse():
  for i in range(len(a)//2):
    temp = a[i]
    a[i] = a[len(a)-1]
    a[len(a)-1] = temp
  return a

def Myindex(n):
  for i in range(len(a)):
    if(a[i] == n):
      return i
  return None

def Myinsert(location, n):
  a.append('')
  for i in range(len(a)):
    if(i < location):
      a[((len(a) - 1) - i)] = a[((len(a) - 1) - i) - 1]
  a[location] = n
  return a

def Myremove(n):
  if(Myindex(n) != None):
    del(a[Myindex(n)])
  return a

def Mycount(n):
  count = 0
  for i in range(len(a)):
    if(a[i] == n):
      count = count + 1
  return count

def Mylen():
  return len(a)

print("원본 배열: ",a)
print("Myappend(이): ",Myappend('이'))
print("Mypop(): ",Mypop())
print("Myreverse(): ",Myreverse())
print("Myindex(주): ",Myindex('주'))
print("Myindex(없는거): ",Myindex('신'))
print("Myinsert(2, 미뇽): ",Myinsert(2, '미뇽'))
print("Myremove(티노): ",Myremove('티노'))
print("Myremove(없는거): ",Myremove('석주'))
print("Mycount(미뇽): ",Mycount('미뇽'))
print("Mylen: ",Mylen())
```

```python
my_dict = {"name": "Alice", "age": 30, "city": "New York"}

# 딕셔너리 출력
print("딕셔너리:", my_dict)

# 딕셔너리 값에 접근
print("이름:", my_dict["name"])
print("나이:", my_dict["age"])
print("도시:", my_dict["city"])

# 새로운 항목 추가
my_dict["email"] = "alice@example.com"

# 딕셔너리 출력
print("업데이트된 딕셔너리:", my_dict)

# 특정 키에 해당하는 항목 삭제
del my_dict["age"]

# 딕셔너리 출력
print("age 키 삭제 후:", my_dict)

# pop() 메서드를 사용하여 삭제하고 값 반환
removed_value = my_dict.pop("city")

# 딕셔너리와 삭제된 값 출력
print("city 키 삭제 후:", my_dict)
print("삭제된 값:", removed_value)
```

```python
#세트는 항목의 중복을 허용하지 않으며, 순서가 없기 때문에 인덱스를 사용하여 항목에 접근할 수 없습니다.
#일반적으로 세트는 항목의 존재 여부를 확인하거나 중복을 제거하기 위해 사용됩니다.

# 세트 생성
my_set = {1, 2, 3, 4, 5}

# 세트 출력
print("세트:", my_set)

# 세트에 항목 추가
my_set.add(6)

# 세트 출력
print("업데이트된 세트:", my_set)

# 세트에서 항목 제거
my_set.remove(3)

# 세트 출력
print("3이 제거된 세트:", my_set)
```

```python
#연결리스트

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def print_list(self):
        current = self.head
        while current:
            print(current.data, end=" ")
            current = current.next
        print()

# 연결 리스트 생성
my_list = LinkedList()

# 요소 추가
my_list.prepend(3)
my_list.prepend(5)
my_list.prepend(7)

# 리스트 출력
my_list.print_list()
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        print("Stack:", self.items)

# 스택 생성
stack = Stack()

# 스택에 항목 추가
stack.push(1)
stack.push(2)
stack.push(3)

# 스택에서 항목 제거
print(stack.pop())  # 출력: 3
print(stack.pop())  # 출력: 2

# 스택이 비어있는지 확인
print("스택이 비어있는가?", stack.is_empty())  # 출력: False
stack.display()
```

```python
from collections import deque

class Queue:
    def __init__(self):
        self.items = deque()

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        return self.items.popleft()

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        print("Queue:", list(self.items))

# 큐 생성
queue = Queue()

# 큐에 항목 추가
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

# 큐에서 항목 제거
print(queue.dequeue())  # 출력: 1
print(queue.dequeue())  # 출력: 2

# 큐가 비어있는지 확인
print("큐가 비어있는가?", queue.is_empty())  # 출력: False
queue.display()
```
